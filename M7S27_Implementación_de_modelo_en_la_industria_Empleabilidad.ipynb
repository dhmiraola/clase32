# Preparando un modelo para producción

Los proyectos de ML son sistemas dinámicos que dependen en gran medida de los datos de entrada. A diferencia del software convencional, la mayoría de ellos se degradan con el tiempo o se vuelven cada vez más irrelevantes. Este problema también se conoce como obsolescencia del modelo. Algunos de los problemas que pueden ocurrir después de implementar el modelo ML en producción son:

- **Desviación de datos**: cuando la distribución de la característica de entrada, es decir, las variables independientes, cambian drásticamente de lo que el modelo ha visto en el entrenamiento.
- **Desviación del modelo o concepto**: cuando las propiedades de las variables objetivo, es decir, las variables dependientes, cambian sin que se modifiquen las características de entrada.
- **Sesgo de servicio de capacitación**: cuando el modelo en producción no tiene el mismo rendimiento que el modelo en entrenamiento.
- **Errores técnicos** y cosas similares.

Para detectar estos problemas a tiempo y tomar medidas, necesitaríamos implementar una estrategia de monitoreo relevante.

how-to-make-sure-your-new-model-is-better-than-the-one-in-poduction-1.webp


Además de la supervisión, un antídoto contra la obsolescencia del modelo es la estrategia de reentrenamiento que se implemente. El tiempo de reentrenamiento del modelo depende del caso de uso comercial, pero, en general, existen cuatro enfoques diferentes:

- **Según el intervalo de tiempo:** entrenar el modelo todos los días, semanas, meses o similar.
- **Basado en el rendimiento:** entrenarlo cuando el rendimiento del modelo cae por debajo de un umbral predefinido.
- **Basado en cambios de datos:** activar el entrenamiento después de cambios significativos en los datos o después de introducir nuevas funciones.
- **Volver a entrenar manualmente** el modelo por alguna otra razón.

Saber qué modelo es «mejor» es una tarea muy desafiante. Un gran reto que surge de inmediato es el sobreajuste. Es el problema cuando el modelo ML se ajusta demasiado a los datos de entrenamiento, lo que conduce a un rendimiento deficiente en los nuevos datos. Esto puede suceder incluso para los profesionales experimentados en aprendizaje automático, ya que no existe una frontera clara entre un sobreajuste y un buen ajuste.

Un desafío más es elegir la métrica adecuada para la evaluación del modelo, que tendrá en cuenta todas las necesidades comerciales. Por ejemplo, en 2009 Netflix otorgó un premio de 1 millón de dólares a un equipo de desarrolladores por mejorar 10 % el algoritmo de recomendación de Netflix. Al final, nunca usaron esta solución porque era demasiado complicada para implementarla en producción y el costo de ingeniería no valía la pena.

## Métricas de evaluación del modelo

Antes de implementar un nuevo modelo en producción, debemos asegurarnos de que tenga un rendimiento óptimo. Hay muchas métricas de evaluación diferentes que se pueden usar para comparar modelos, y elegir la correcta es algo crucial. Aquí tienes los más populares.

## Métricas de clasificación

Cuando se trata de métricas de clasificación, los principales factores que se deben considerar para elegirlas son:

- **El número de clases**: clasificación binaria o multiclase.
- **El número de muestras por clase**: ¿tenemos un conjunto de datos equilibrado?
- **Caso de uso comercial**: por ejemplo, equilibrio entre precisión y recuperación en función del caso de uso comercial.

La métrica de clasificación más utilizada es la **precisión**. Para conjuntos de datos más desequilibrados, se utilizan métricas como F1, precisión y recuperación. Todos ellos y muchos más se pueden calcular a partir de la matriz de confusión. Para la clasificación multiclase, se utilizan métricas similares con fórmulas ligeramente diferentes. Para utilizar la probabilidad de la clase predicha, se utilizan métricas como ROC y AUC.

## Métricas de regresión

Las métricas de regresión, por lo general, calculan algún tipo de distancia entre los valores reales predichos y los reales, lo que se expresa como un error. Las métricas de regresión más utilizadas son:

- Error cuadrático medio (MSE)
- Error absoluto medio (MAE)
- Error cuadrático medio de la raíz (RMSE)
- Error porcentual absoluto medio (MAPE)
- R-cuadrada







# Pruebas A/B para ML

Las pruebas A/B, en el nivel más básico, son una forma de comparar dos versiones de un producto en un intento de comprender qué funciona mejor. Para hacer esto primero debemos entender qué es un intervalo de confianza:

## Intervalo de confianza

Es un rango de valores que probablemente representen la media de la población con un cierto grado de confianza. El nivel de confianza más común es del 95 % y está ligado libremente por un límite inferior y uno superior. Esto, en esencia, significaría que puede estar 95 % seguro de que la media de la población se ubicaría entre los límites superior e inferior si la misma técnica de muestreo se repitiera varias veces.

## Prueba de hipótesis

Una hipótesis puede definirse como una suposición sobre algo en el mundo que puede ser probada mediante un experimento o mediante observaciones. En términos generales, una declaración de hipótesis se vería así:

## Puntuación Z (Z Score)

La puntuación Z es una medida estadística que nos dice qué tan lejos está un punto de datos del resto del conjunto de datos. En términos más técnicos, la puntuación Z indica cuántas desviaciones estándar de una observación dada son de la media.

La puntuación Z es una medida paramétrica y toma dos parámetros: media y desviación estándar.

\begin{align}
Z-score = \frac{x - media}{Desviación estandar}
\end{align}

Una vez que calculamos estos dos parámetros, es fácil encontrar el puntaje Z de un punto de datos.

Basta de teoría, vamos a la práctica. Para el ejemplo usaremos el siguiente *dataset*: https://drive.google.com/file/d/1YFkTZkw9b62DLDfw3UwNDFf-lIfeJHwb/view?usp=share_link


````python
# Importamos bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.power import NormalIndPower
from scipy.stats import norm

from google.colab import drive
drive.mount('/content/drive')
````

````python
# Cargamos el dataset
df = pd.read_csv('/content/drive/MyDrive/Ucamp/Semana 27/redesign.csv')
df.info()
````

````python
# formamos un nuevo punto de datos para comprender el tratamiento y las nuevas imágenes.
groups = df.groupby(['treatment', 'new_images'])

groups['converted'].plot.kde()
plt.legend()
plt.show()

results = groups.agg({'converted':['count', 'mean', 'sum']})
results
````

````python
# Definimos el nivel de significación para la aceptación:
alpha = 0.05

# Preprocesamos los parámetros para prepararlos para la prueba de puntuación z
def preprocess(experiment, control):
    mean_one = experiment["converted"].mean()
    sum_one = experiment["converted"].std()
    count_one = experiment["converted"].count()

    mean_two = control["converted"].mean()
    sum_two = control["converted"].std()
    count_two = control["converted"].count()

    return [mean_one, sum_one, count_one], [mean_two, sum_two, count_two]

# Calculamos los puntajes z en los grupos de experimento y control
def zscore_test(experiment, control, val=0):
    exp, ctrl = preprocess(experiment, control)
    return (exp[0] - ctrl[0] - val)/(np.sqrt(exp[1]**2/exp[2] + ctrl[1]**2/ctrl[2]))

# Calculamos el límite inferior de los intervalos de confianza para la prueba
def confidence(experiment, control, Z):
    exp, ctrl = preprocess(experiment, control)
    lower_interval = (exp[0] - ctrl[0]) - Z * (np.sqrt(exp[1]**2/exp[2] + ctrl[1]**2/ctrl[2]))
    return lower_interval

# Calculamos el valor alfa del puntaje z para rechazar la hipótesis nula:
z_alpha = norm.ppf(1-alpha)
print("z_alpha: {}".format(z_alpha))
results[('converted', 'z_score')] = ""

group_ctrl = groups.get_group(("no", "no"))

for key in list(groups.groups.keys())[1:]:
    group_exp = groups.get_group(key)
    zscore = zscore_test(group_exp, group_ctrl)
    results.loc[key, ('converted', 'z_score')] = zscore

results
````

````python
# Ahora podemos descartar la hipótesis nula para el grupo (sí, no):

control = groups.get_group(("no", "no"))
experiment = groups.get_group(("yes", "no"))

lower = np.round(confidence(experiment, control, z_alpha) * 100, 2)

print("Con 95 % de confianza podemos afirmar que la mejora del diseño (SÍ, NO) con respecto al grupo control fue {}%.".format(lower))

print("\nPara que (1) sea cierto, la probabilidad de cometer un error de tipo I al llegar a la conclusión anterior es de aproximadamente {}%.".format(alpha*100))

print("\nPara que (1) sea cierto, la probabilidad de cometer un error de tipo I al llegar a la conclusión anterior es de aproximadamente {}%.".format(np.round( (1 - results.loc[("yes", "no"), ("converted", "z_score")]) * 100, 2)))
````

A través de nuestras pruebas A/B, concluimos, con 95 % de confianza, que al ver la nueva versión del modelo con imágenes antiguas, las posibilidades  mejoran en al menos 0.56 % en comparación con el diseño anterior. También tomamos en cuenta los errores de Tipo I y Tipo II para anular las posibilidades de un análisis incorrecto.

# Introducción a Flask

Flask es un *microframework* de  Python que permite crear aplicaciones web.

Podemos definir un *framework* como un esquema o un esqueleto para el desarrollo y/o la implementación de una aplicación.

Flask proporciona lo mínimo necesario para hacer funcionar una aplicación web rápidamente, en este caso una web API REST.

## Application Programming Interface (API)

Una interfaz de programación de aplicaciones (API) es un código que permite que dos programas de software se comuniquen. Una API define cómo un desarrollador debe solicitar servicios de un sistema operativo (SO) u otra aplicación, y exponer datos dentro de diferentes contextos y a través de múltiples canales.

Una **API RESTful** es un estilo arquitectónico para una interfaz de programa de aplicación (API) que usa solicitudes HTTP para acceder y usar datos. Estas pueden utilizarse para los tipos de datos GET, PUT, POST y DELETE, que se refieren a la lectura, actualización, creación y eliminación de operaciones relacionadas con los recursos.

## Verbos HTTP

Estas son algunas reglas que te servirán para saber cómo y cuándo debes usar los verbos HTTP en una arquitectura REST. Los verbos HTTP involucrados en un sistema REST son GET, POST, PUT, PATCH y DELETE:

- **GET** se usa para consultar un recurso. Una de las principales características de una petición GET es que no debe causar efectos secundarios en el servidor, ni producir nuevos registros, ni modificar los ya existentes.
- Las peticiones con **POST** son solo para crear recursos nuevos. Cada llamada con POST debería producir un nuevo recurso.
- Los verbos **PUT** y **PATCH** son muy similares, ya que ambos se usan para modificar un recurso existente.
- Por último, **DELETE** es el verbo que se usa para eliminar registros, bien pudiera ser para eliminar un recurso individual.

````python
# Vamos a usar un clasificador de bosque aleatorio:
from sklearn.ensemble import RandomForestClassifier
````

````python
# Cargamos los datos:
df = pd.read_csv('/content/drive/MyDrive/Ucamp/Semana 27/titanic.csv')
df.head(1)
````

````python
# Eliminamos columnas no deseadas:
df.drop(["PassengerId","Name","Cabin","Ticket"],axis=1,inplace=True)
````

````python
# Recuento de valores faltantes en cada columna:
print(df.isnull().sum())
````

````python
# Eliminación de filas con valores faltantes:
df.dropna(subset=['Embarked','Age'],inplace=True)
print(df.isnull().sum())
````

````python
# Codificación One hot encoding para variables categóricas
# Creación Dymmies:
dummies = pd.get_dummies(df.Sex)
dummies2 = pd.get_dummies(df.Embarked)
````

````python
# Agregamos el dataset de los dummies con el dataset original:
new_df = pd.concat([df,dummies,dummies2],axis='columns')
print(type(new_df))
````

````python
# Eliminamos columnas que no usaremos:
new_df.drop(['Sex','Embarked'],axis='columns',inplace=True)
print(new_df.head(10))

new_df.info()
````

````python
# Dividimos el dataset en características y clase target:
x = new_df[new_df.columns.difference(['Survived'])]
y = new_df['Survived']

classifier = RandomForestClassifier()
classifier.fit(x, y)
````

Finalmente, podemos implementar el modelo persistente utilizando un marco web. El siguiente código crea una API REST usando Flask:

``` python
from flask import Flask
from flask import request
import pandas as pd
import joblib


app = Flask(__name__)

@app.route('/predice', methods=['POST'])
def predict():
    json_ = request.json
    query_df = pd.DataFrame(json_, index=[0])
    query = pd.get_dummies(query_df)
    
    classifier = joblib.load('classifier.pkl')
    prediction = classifier.predict(query)

    if prediction[0] :
        return "TRUE: El pasajero pudo haber sobrevivido"
    else :
        return "FALSE: El pasajero pudo NO haber sobrevivido"


if __name__ == "__main__":
    app.run(port=8000, debug=True)

```

# Crear un repositorio


Antes de subirlo a tu portafolio, debes crear un repositorio local de Git. Para ello, escribe dentro de tu terminal y dentro de tu carpeta de proyecto:

``` bash
git init
```

Después necesitarás incluir todos los archivos dentro del repositorio, para ello realiza lo siguiente:

``` bash

```

Para agregar un solo archivo al repositorio:
``` bash
git add <NOMBRE_Y_RUTA_DEL_ARCHIVO
```

Para agregar una carpeta con todos los archivos que contiene:
``` bash
git add <NOMBRE_Y_RUTA_DE_LA_CARPETA>
```

Para agregar todos los archivos del proyecto:
``` bash
git add
```

En la carpeta de tu proyecto se crearon dos carpetas: `venv` y `__pycache__`, las cuales no debes subir al repositorio, ya que se generan automáticamente.

Para excluirlas del proyecto, crea un archivo llamado `.gitignore` e incluye las rutas y nombres de los archivos o carpetas que quieres excluir. Por ejemplo:

```
__pycache__
venv
```

Después necesitas hacer un «corte de caja», es decir, una forma de decir «hasta aquí ya está bien, llegamos a una meta». Esto se llama *commit*. Para hacerla, ejecuta el siguiente comando:
```bash
git commit -m "COMENTARIO_SOBRE_LOS_CAMBIOS"
```

Luego de esto, crea un repositorio de GitHub desde el siguiente *link* -> https://github.com/new

Te aparecerá una pantalla como esta:

Captura de pantalla 2023-01-19 a la(s) 22.54.09.png

Te recomendamos que solo rellenes el nombre del repositorio, ya que así el mismo GitHub te dará la lista de comandos a ejecutar en la terminal para que puedas subir tu proyecto.




Una vez creado el repositorio remoto, debes conectarlo con el repositorio local para que puedas subir los cambios. Esto se logra con el siguiente comando:
``` bash

git remote add origin <DIRECCION_REMOTA_DEL_REPOSITORIO>

```

Por defecto, Git crea la rama máster para los proyectos. Las ramas son versiones del software que pueden estar en paralelo y luego unirse. Por otro lado, GitHub usa la rama *main*, así que habrá que cambiar la rama local, ya que ambas deben coincidir. Esto se hace con el siguiente comando:
``` bash
git branch -M main
```

Por último, sube los cambios al repositorio remoto con el siguiente comando:
``` bash
git push -u origin main
```
